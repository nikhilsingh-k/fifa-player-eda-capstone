## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df_clean = df.copy()
df_clean.drop_duplicates(inplace=True)
df_clean.fillna(method='ffill', inplace=True)
df_clean['Joined'] = pd.to_datetime(df_clean['Joined'], errors='coerce')
df_clean['Contract Valid Until'] = pd.to_numeric(df_clean['Contract Valid Until'], errors='coerce')

df_clean.info()
````

### What all manipulations have you done and insights you found?

Manipulations Performed

Several data manipulation steps were carried out to make the dataset analysis-ready:

Created a copy of the original dataset to preserve raw data.
Removed duplicate records to avoid repeated information.
Handled missing values using forward fill techniques to maintain continuity.
Converted date columns such as Joined into datetime format for proper analysis.
Converted contract-related fields into numeric format for consistency.
Verified data types to ensure compatibility with statistical operations and visualizations.
Insights Found
The dataset contains a rich combination of numerical, categorical, and date-related variables.
A small number of missing values were identified and handled appropriately.
Duplicate records were minimal, indicating good dataset quality.
Date conversion enables better time-based analysis such as player tenure.
Proper formatting of variables improves the accuracy of visualizations and insights.

Overall, these manipulations improved data reliability and prepared the dataset for meaningful exploratory analysis.

#### Chart - 1

````python
# Chart - 1 visualization code

plt.figure(figsize=(10,6))
sns.histplot(df_clean['Age'], bins=20, kde=True)

plt.title("Age Distribution of FIFA Players")
plt.xlabel("Age")
plt.ylabel("Number of Players")

plt.show()
````

##### 1. Why did you pick the specific chart?

This chart was chosen to analyze the age distribution of football players in the dataset. Age is a crucial factor in player performance, career growth, and market value.

##### 2. What is/are the insight(s) found from the chart?

The majority of players are concentrated between the ages of 20 and 30, which represents the prime years of professional football careers.

There are fewer players at very young or older ages, indicating that clubs generally focus on athletes within their peak performance range.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the gained insights can create a positive business impact.

By understanding that most players fall within the 20–30 age range, football clubs can make more informed recruitment decisions. Players in this age group are generally at their peak physical and technical performance, making them valuable assets for both short-term success and competitive consistency.

This insight helps organizations optimize investments by targeting players who can contribute effectively while maintaining strong market value.

However, there are also insights that may lead to negative growth if not managed carefully.

For example, over-reliance on players in the peak-age category may result in higher wage costs and reduced long-term sustainability. Clubs that invest heavily only in experienced players may miss opportunities to develop younger talent with higher potential and future resale value.

Additionally, neglecting younger age groups can weaken succession planning, leading to performance gaps when senior players retire or decline.

Therefore, while focusing on prime-age players supports immediate results, a balanced recruitment strategy that includes young prospects is essential for sustainable growth and long-term competitiveness.
